Bivariate analysis - 2 variables at a time

multivariate analysis - multiple variables at a time

## 1. SCATTER PLOT(NUMERICAL - NUMERICAL)

```python
#Bivariate Analysis
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'])
plt.show()
```

```python
#Multivariate analyssis
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="size",size="size",palette="deep",sizes=(20,200),legend="full")
plt.show()
```

## BAR PLOT(NUMERIC - CATEGORICAL)

```python
#multivariate Analysis
```

```python
sns.boxplot(data=titanic,x="Sex",y="Age",hue="Survived")
plt.show()
```

## DIST PLOT(NUMERIC - CATEGORICAL)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
plt.show()
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
plt.show()
```

## HEATMAP (CATEGORICAL - CATEGORICAL)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived']) #frequency table
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
plt.show()
```

```python
titanic.groupby("Pclass").mean()['Survived']*100
```

```python
(titanic.groupby("Pclass").mean()['Survived']*100).plot(kind="bar")
plt.show()
```

```python
titanic.groupby("Embarked").mean()['Survived']*100
```

```python
(titanic.groupby("Embarked").mean()['Survived']*100).plot(kind="bar")
plt.show()
```

Southampton, Cherbourg, and Queenstown

```python
titanic.groupby("Pclass").Survived.value_counts()
```

## CLUSTERMAP(CATEGORICAL-CATEGORICAL) 

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
plt.show()
```
